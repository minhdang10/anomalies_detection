```python
%%snowflakes vs

SELECT 
    P.FACILITYNAME, P.CREATIONTIME, P.MACHINEID, P.LENGTH, P.WIDTH, P.HEIGHT, P.WEIGHT, 
    B.BARCODE, PF.MERCHANT, PF.MERCHANT_ORG_KEY 
FROM 
    FDR_DB.FDR_AUTOMATION.PARCEL P
INNER JOIN 
    FDR_DB.FDR_AUTOMATION.BARCODE B
ON 
    B.PARCELKEY = P.PARCELKEY 
INNER JOIN 
    FDR_DB.FDR_EWS.PARCEL_FACILITY PF
ON 
    P.FACILITYNAME = PF.FACILITY_NAME 
AND
    B.BARCODE = PF.PRIMARY_BARCODE
WHERE
    PF.MERCHANT = 'Victoria Secret'
AND
    P.FACILITYNAME = 'DFW4';
```
